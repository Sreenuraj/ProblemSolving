# Recrussive Solution

```python
def fib(num):
    if num == 1 or num == 2:
        result = 1
    else:
        result = fib(num-1) + fib(num-2)
    return result
```

```python
%%time
fib(10)
```

# Memoized solution

```python
memo = {}
def fib(num):
    if num in memo:
        return memo[num]
    if num == 1 or num == 2:
        result = 1
    else:
        result = fib(num-1) + fib(num-2)
    memo[num] = result
    return result
```

```python
%%time
fib(1000)
```

# Bottom up approch

```python
def fib_bottom_up(num):
    if num == 1 or num == 2:
        return 1
    bottom_up = [None] * (num+1)
    bottom_up[1] = 1
    bottom_up[2] = 1
    for i in range (3, num+1):
        bottom_up[i] = bottom_up[i-1] + bottom_up[i-1]
    return bottom_up[num]
```

```python
%%time
fib_bottom_up(1000)
```

```python
%%time
fib_bottom_up(10000)
```
